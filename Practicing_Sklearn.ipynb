```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('Real_life_example.csv')
```

```python
data.head()
```

```python
data.describe(include='all')
```

```python
# check for missing values
data.isnull().sum()
```

```python
data.dropna(axis=0,inplace=True)
```

```python
data.describe(include='all')
```

```python
# View the Outliers
sns.boxplot(data['Price'])
```

```python
p = data['Price'].quantile(0.75)
```

```python
d = data[data['Price']<p]
```

```python
d.shape
```

```python
sns.boxplot(d['Price'])
```

```python
sns.boxplot(d['Mileage'])
```

```python
m = d['Mileage'].quantile(0.99)
d2 = d[d['Mileage']<m]
```

```python
d2.shape
```

```python
sns.distplot(d2['EngineV'])
```

```python
d3 = d2[d2['EngineV']<6.5]
```

```python
sns.distplot(d3['EngineV'])
```

```python
d3.shape
```

```python
sns.boxplot(d3['Year'])
```

```python
q = d3['Year'].quantile(0.01)
d4 = d3[d3['Year']>q]
```

```python
sns.boxplot(d4['Year'])
```

```python
d4.shape
```

```python
d4.info()
```

```python
d4.describe(include='all')
```

```python
cat = [col for col in d4.columns.values if d4[col].dtype == 'object']
d_cat = d4[cat]
```

```python
d_num = d4.drop(d_cat,axis=1)
```

```python
d_num.info()
```

```python
num_s = ((d_num - d_num.mean()/d_num.max() - d_num.min()))
```

```python
num_s.head()
```

```python
cat_dummies = pd.get_dummies(d_cat,drop_first=True,dtype=int)
```

```python
cat_dummies.head()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

sns.scatterplot(x='Year',y='Price',data=num_s, ax=axes[0,0])
sns.scatterplot(x='Mileage',y='Price',data=num_s, ax=axes[0,1])
sns.scatterplot(x='EngineV',y='Price',data=num_s, ax=axes[1,0])

plt.show()
```

```python
full_d = pd.concat([num_s,cat_dummies],axis=1)
```

```python
full_d.corr()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
target = full_d['Price']
ff = full_d.drop('Price',axis=1)
ff.info()
```

```python
x_train,x_test,y_train,y_test = train_test_split(ff,target,test_size=0.2,random_state=0)
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
y_hat = lm.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('Targets (y_train)',size=18)
plt.ylabel('Predictions (y_hat)',size=18)
```

```python
lm.score(x_train,y_train)
```

```python
y_hat_test = lm.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test, alpha=0.2)
plt.xlabel('Targets (y_test)',size=18)
plt.ylabel('Predictions (y_hat_test)',size=18)
plt.show()
```

```python
df_pf = pd.DataFrame(y_hat_test, columns=['Prediction'])
df_pf.head()
```

```python
y_test = y_test.reset_index(drop=True)
df_pf['target'] = y_test
```

```python
df_pf
```

```python
df_pf['Residual'] = df_pf['target'] - df_pf['Prediction']
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['target']*100)
df_pf
```

```python
pd.options.display.max_rows = 999
# Moreover, to make the dataset clear, we can display the result with only 2 digits after the dot 
pd.set_option('display.float_format', lambda x: '%.2f' % x)
# Finally, we sort by difference in % and manually check the model
df_pf.sort_values(by=['Difference%'])
```
